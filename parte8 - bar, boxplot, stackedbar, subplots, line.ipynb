## Instalando e Carregando os Pacotes

```python
# Imports
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline 
```

```python
# Improtando dados
dados = pd.read_csv("dados/dataset7.csv")
```

```python
dados.shape
```

> DICA: para o pandas exibir todas as colunas do dataset, utilizamos o comando abaixo

```python
pd.options.display.max_columns = dados.shape[1]
```

```python
dados.head()
```

> Procedimento para remover ' ' e ',' da variável 'income'

```python
dados.rename({' Income ':'Income'}, axis = 1, inplace = True)
dados['Income'] = dados['Income'].str.replace('$','').str.replace(',','').astype(float)
```

```python
# Verificando valores nulos
dados.isnull().sum()
```

```python
# Substituindo os nulos pela média 
dados['Income'].fillna(dados['Income'].mean(), inplace = True)
```

```python
dados.info()
```

```python
# Convertendo Dt_Customer em datetime
dados['Dt_Customer'] = pd.to_datetime(dados['Dt_Customer'])
```

```python
# Calculando idade do cliente
dados['Customer_age'] = dados['Dt_Customer'].dt.year - dados['Year_Birth']
```

```python
# Calculando GRUPO etário (diferente de faixa etária) em: Jovem - Adulto - Idoso
dados['Grupo_Etario_Clientes'] = pd.cut(dados.Customer_age, bins = [0,25,50,150], labels = ['Jovem', 'Adulto', 'Idoso'])
```

```python
# Total de compras (somando categprias)
dados['Total_Purchases'] = dados['NumDealsPurchases']+dados['NumWebPurchases']+dados['NumCatalogPurchases']+dados['NumStorePurchases']
```

```python
# Verificando se tem crianças ou adolescentes em casa e criamos uma variável para isso
dados['Children'] = dados['Kidhome']+dados['Teenhome']
```

```python
dados.head()
```

#### 01- Gráfico de barras com: relação entre income (salário) e país com base no grupo etário

```python
plt.figure(figsize = (16,12))

sns.barplot(x = 'Country', 
            y = 'Income', 
            hue = 'Grupo_Etario_Clientes', 
            data = dados, 
            estimator = np.mean, 
            palette = 'autumn')

plt.title('Salário x País', fontsize = 18)
plt.xlabel('País', fontsize  = 12)
plt.ylabel('Salário', fontsize = 12)
```

> Análise: nota-se por exemplo qu ena Alemanha os jovens são os que mais fazem compras

#### 02 - Boxplot com a distribuição da variável de total de compras com base na idade

```python
plt.figure(figsize = (12,9))
sns.boxplot(x = 'Grupo_Etario_Clientes', y = 'Total_Purchases', data = dados, palette = 'autumn')
plt.title('Distribuição do Total de Compras Pela Idade', fontsize = 18)
plt.xlabel('Grupo Etário', fontsize = 12)
plt.ylabel('Total de Compras', fontsize = 12)
```

> Análise: nota-se que os idosos comparam mais que adultos e jovens, mas nota-se outliers em adultos que compraram muito acima da média

#### 03 - Boxplot com a distribuição da variável de total de compras com base no país

```python
plt.figure(figsize = (12,9))
sns.boxplot(x = 'Country', y = 'Total_Purchases', data = dados, palette = 'Blues_r')
plt.title('Total de Compras com Base no País', fontsize = 18)
plt.xlabel('País', fontsize = 12)
plt.ylabel('Total de Compras', fontsize = 12)
```

> Análise: agora temos mais países que grupos etários, então o uso de paleta de cores pode nos auxiliar com a evolução
> Quase todos os países as pessoas tiveram o mesmo padrão, em exceção no México que ficou bem abaixo

#### 04 - Stackedbar (Gráfico de barras empilhadas) com a relação entre número de crianças em casa e número de produtos comprados, com a legenda contendo o tipo de produto

> Gráfico de interpretação e elaboração mais avançada, onde:
- Criamos um grupo de produtos com os nomes extraídos das variváveis, pois não temos uma variável específica (isso poderia ser feito em etapas anteriores), ou seja só pegamos os nomes
- Criamos uma lista de cores para usar no laço for
- No laço for:
    1. navegamos pelos produtos e pelas cores, descompactando com a função zip()
    2. Também no laço criamos a sobreposição dos gráficos usando as informações
    3. O 'bottom' é usado para poder construir os gráficos de baixo para cima 
    4. A altura é definida com o contador 'p'
    5. Em plt.legend() usamos o replace() para excluir o 'Mnt' e 'Products' para deixar somente o nome do produto

```python
plt.figure(figsize = (16,7))

# Criamos um array com os tipos de produtos
products = ['MntWines', 
            'MntFruits',
            'MntMeatProducts', 
            'MntFishProducts', 
            'MntSweetProducts',
            'MntGoldProds']

color = ['purple', 'green', 'gold', 'blue', 'red', 'brown']

for i, p, c in zip(range(1,len(products)+1), products, color):
    
    if i != 6:
        bottom = dados[products[i]]
        plt.bar(x = dados['Children'], height = dados[p], color = c, bottom = bottom)
    else:
        plt.bar(x = dados['Children'], height = dados[p], color = c)
        
plt.xlabel('Crianças em Casa')
plt.ylabel('Número de Produtos Comprados')
plt.title('Número de Produtos Comprados x Crianças em Casa', fontsize = 15)

plt.legend([p.replace('Mnt','').replace('Products','') for p in products])
plt.xticks(dados['Children'].unique())
```

> Análise: quem não tem criança, compra muita carne. Quem tem 2 ou 3 crianças compra mais vinho<br>
> Apesar de ser um ótimo gráfico, pode causar problemas de interpretação

#### 05 - Subplots com gráficos de barras mostrando a relação entre número de crianças em casa e o número de produtos comprados. Cada subplot representa um tipo de produto.

> Utilizamos basicamente o gráfico anterior para produzir subplots, usando um laço for para construir os plots de forma separada

```python
products = ['MntWines', 
            'MntFruits',
            'MntMeatProducts', 
            'MntFishProducts', 
            'MntSweetProducts',
            'MntGoldProds']

row, col = 2, 3
p = 0

fig, ax = plt.subplots(row, col, figsize = (17,8))

for i in range(row):
    for j in range(col):
        plot = sns.barplot(x = dados['Children'], y = dados[products[p]], palette = 'rainbow', ax = ax[i,j])
        plot.set_title('Produto: ' + products[p].replace('Mnt',''))
        p += 1
        
fig.tight_layout()
```

#### 06 - Subplots com gráficos de barras mostrando a relação entre estado civil e o número de produtos comprados. Cada subplot representa um tipo de produto.

> Os 2 gráficos posteriores usam a mesma lógica dos gráficos anteriores mas com variáveis diferentes

```python
products = ['MntWines', 
            'MntFruits',
            'MntMeatProducts', 
            'MntFishProducts', 
            'MntSweetProducts',
            'MntGoldProds']

row, col = 2, 3
p = 0

fig, ax = plt.subplots(row, col, figsize = (17,8))

for i in range(row):
    for j in range(col):
        plot = sns.barplot(x = dados['Marital_Status'], y = dados[products[p]], palette = 'rainbow', ax = ax[i,j])
        plot.set_title('Produto: ' + products[p].replace('Mnt',''), fontsize = 15)
        p += 1
        
fig.tight_layout()
```

#### 07 - Stackedbar (Gráfico de barras empilhadas) com a relação estado civil e o número de produtos comprados, com a legenda contendo o tipo de produto

```python
plt.figure(figsize = (16,7))

products = ['MntWines', 
            'MntFruits',
            'MntMeatProducts', 
            'MntFishProducts', 
            'MntSweetProducts',
            'MntGoldProds']

color = ['purple','green','gold','blue','red','brown']

for i, p, c in zip(range(1, len(products)+1), products, color):
    
    if i != 6:
        bottom = dados[products[i]]
        plt.bar(x = dados['Marital_Status'], height = dados[p], color = c, bottom = bottom)
    else:
        plt.bar(x = dados['Marital_Status'], height = dados[p], color = c)
        
plt.xlabel('Estado Civil')
plt.ylabel('Número de Produtos Comprados')
plt.title('Número de Produtos Comprados x Estado Civil', fontsize = 15)

plt.legend([p.replace('Mnt','').replace('Products','').replace('Prods','') for p in products])
plt.xticks(dados['Marital_Status'].unique())
```
