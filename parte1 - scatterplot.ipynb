## VISUALIZAÇÕES COM SCATTER PLOT

> PROBLEMA DE NEGÓCIO:
- O objetivo é compreender a relação entre as variáveis de operação de um restaurante
- **Dataset**: dataset1

> Esse gráfico é muito útil quando se deseja investigar se existe uma relação linear entre as variáveis, ou seja, se uma delas aumenta, a outra também aumenta (ou diminui)

### Instalando e Carregando os Pacotes

```python
# Imports
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dados = pd.read_csv("dados/dataset1.csv")
dados.head(10)
```

```python
# Tamanho (linhas x colunas) do dataset
dados.shape
```

```python
# Tipos de dados
dados.dtypes
```

```python
# Análise estatística
dados.describe()
```

```python
# Verificando ocorrência de dados nulos
dados.isnull().sum()
```

### VISUALIZAÇÕES

##### 01 - Scatter Plot com a relação entre duas variáveis

```python
plt.figure(figsize = (20, 8))
sns.relplot(            # relplot() permite criar um plot relacional
    x = 'total_bill',   # valor da conta
    y = 'tip',          # gorjeta
    data = dados
)
```

#### 02 - Scatter Plot com a relação entre duas variáveis e com pontos coloridos para diferenciar categorias(fumante ou não) de uma terceira variável

```python
plt.figure(figsize = (20, 8))
sns.relplot(            # relplot() permite criar um plot relacional
    x = 'total_bill',   # valor da conta
    y = 'tip',          # gorjeta
    hue = 'smoker',     # se é fumante ou não
    data = dados
)
```

#### 03 - Scatter Plot com a relação entre duas variáveis, pontos coloridos para diferenciar categorias e uma terceira variável com símbolo diferente

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    hue = 'smoker',
    style = 'smoker',  # um novo formato para a informação   
    data = dados
)
```

#### 04 - Scatter Plot com a relação entre duas variáveis, pontos coloridos para diferenciar categorias, uma terceira variável (fumante ou não), com símbolos diferentes indicando se o cliente foi almoçar ou jantar

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    hue = 'smoker',
    style = 'time',  # mais uma informação adicionada
    data = dados
)
```

#### 05 - Scatter Plot com a relação entre duas variáveis e pontos coloridos para diferenciar o 'size' (número de pessoas na mesa)

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    hue = 'size',
    data = dados
)
```

#### 06 - Scatter Plot com a relação entre duas variáveis e pontos coloridos para diferenciar o 'size' (número de pessoas na mesa), mas com formatação de cor por tamanho (quanto maior, mais escuro)

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    hue = 'size',
    palette = "ch:r=-.5,l=.75",
    data = dados
)
```

#### 07 - Scatter Plot com a relação entre duas variáveis e os pontos com tamanhos definidos pelas categorias de uma terceira variável no caso 'size'

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    size = 'size',   # tamanho dos pontos
    data = dados
)
```

> Essa visualização pode ser confusa, pois o tamanho de casa ponto fica confuso de ser analisado

#### 08 - Scatter Plot com a relação entre duas variáveis e os pontos com tamanhos definidos pelas categorias de uma terceira variável no caso 'size', mas com escala maior (sizes)

```python
plt.figure(figsize = (20, 8))
sns.relplot(            
    x = 'total_bill',   
    y = 'tip',          
    size = 'size',
    sizes = (15, 200),
    data = dados
)
```
